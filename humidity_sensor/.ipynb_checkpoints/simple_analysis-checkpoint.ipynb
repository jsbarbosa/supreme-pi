```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
%matplotlib inline
```

```python
def filtering(coeff, number = 1):
    temp = coeff.copy()
    results = np.zeros_like(temp)
    for i in range(2*number):
        pos = temp.argmax()
        results[pos] += temp[pos]
        temp[pos] = 0
        
    return results
```

```python
data = np.genfromtxt("DHT_results.csv", delimiter = ",")

data = data[2549:]
while True:
    pos = data[:,2].argmax()
    if np.isnan(data[pos,2]):
        data[pos] = data[pos-1] 
    else:
        if data[pos,2] > 100:
            data[pos] = data[pos-1]
        else:
            break
            
temperature = data[:,1]
humidity = data[:,2]
```

```python
plt.plot(temperature)
coeff = np.fft.rfft(temperature)
filtered = np.real(np.fft.irfft(filtering(coeff, 1)))

plt.plot(filtered, lw=2)
plt.ylabel("Temperature (C)")
plt.show()
```

```python
plt.plot(humidity)
coeff = np.fft.rfft(humidity)
filtered = np.real(np.fft.irfft(filtering(coeff, 3)))

plt.plot(filtered, lw=2)
plt.ylabel("Humidity (%)")
plt.show()
```
